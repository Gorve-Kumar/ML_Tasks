```python
import numpy as np
import matplotlib.pyplot as plt
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

x, y = mglearn.datasets.make_wave(n_samples=100)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Wave Dataset")
plt.show()

# IT HAS GOT 100 SAMPLES. USED FOR LINEAR REGRESSION
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print("R2 score:", r2_score(y_test, y_pred))
print("Mean Relative Error (%):", np.mean(np.abs((y_test-y_pred)/y_test))*100)

# Plot the predicted values against the true values
plt.scatter(x_test, y_test, label='True')
plt.plot(x_test, y_pred, color='r', label='Predicted')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Linear Regression on Wave Dataset')
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=3)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print("R2 score:", r2_score(y_test, y_pred))
print("Mean Relative Error (%):", np.mean(np.abs((y_test-y_pred)/y_test))*100)

# Plot the predicted values against the true values
plt.scatter(x_test, y_test, label='True')
plt.plot(x_test, y_pred, color='r', label='Predicted')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('KNN Regression on Wave Dataset')
plt.legend()
plt.show()
```
