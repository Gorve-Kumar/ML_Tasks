```python
inp = df.iloc[:,1:5]
inp
```

```python
out = df.iloc[:,-1]
out
```

```python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

from sklearn.neighbors import KNeighborsClassifier

from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
inp_train, inp_test, out_train, out_test= train_test_split(inp,out,test_size=0.2)
```

```python
scaler=StandardScaler()
inp_train=scaler.fit_transform(inp_train)
inp_test=scaler.fit_transform(inp_test)
```

```python
inp_test
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(inp_train,out_train)
```

```python
out_predict=knn.predict(inp_test)
print(metrics.accuracy_score(out_test,out_predict)*100)
print(confusion_matrix(out_test,out_predict))
```
